```python
# Import necessary libraries
import pandas as pd

try:
    import numpy as np
    print("NumPy imported successfully.")
    # You can add any additional NumPy-dependent code here
except ImportError as e:
    print(f"Ignored ImportError: {e}")
    # Fallback or additional handling can go here if needed

# Continue with the rest of your code
print("Continuing with the rest of the code...")
# Add your main code logic here


# Load the dataset
data = pd.read_csv('Alphabets_data.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values and general information about the data
data.info()
```

```python
# Encode the target variable 'letter' using Label Encoding
from sklearn.preprocessing import LabelEncoder

# Separate features and target variable
X = data.drop(columns='letter')
y = data['letter']

# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Normalizing the feature set
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Building the model
model = Sequential()
model.add(Input(shape=(X_train.shape[1],)))  # Define input layer explicitly
model.add(Dense(64, activation='relu'))  # First hidden layer
model.add(Dense(32, activation='relu'))  # Additional hidden layer
model.add(Dense(len(label_encoder.classes_), activation='softmax'))  # Output layer with softmax for multi-class classification

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

# Define a function to create the model (required for KerasClassifier)
def create_model(optimizer='adam', neurons=64, activation='relu'):
    model = Sequential()
    model.add(Dense(neurons, input_shape=(X_train.shape[1],), activation=activation))
    model.add(Dense(neurons // 2, activation=activation))
    model.add(Dense(len(label_encoder.classes_), activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Wrap Keras model for use in GridSearchCV
model = KerasClassifier(build_fn=create_model, epochs=20, batch_size=32, verbose=0)

# Define grid of hyperparameters
param_grid = {
    'optimizer': ['adam', 'sgd'],
    'neurons': [32, 64, 128],
    'activation': ['relu', 'tanh'],
    'batch_size': [16, 32],
    'epochs': [10, 20]
}

# Grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train, y_train)
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
# Get best hyperparameters and print them
print(f"Best Hyperparameters: {grid_result.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_result.best_score_:.4f}")
```

```python
from sklearn.metrics import classification_report

# Predict and evaluate performance
y_pred = model.predict(X_test).argmax(axis=1)
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```
